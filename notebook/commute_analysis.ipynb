```python
def estimate_commute(home, office):
    """
    Simulates commute time (minutes)
    - based on string length difference
    - adds randomness to simulate traffic
    """

    base = abs(len(home) - len(office)) * 1.5
    traffic = random.uniform(10, 40)

    return round(base + traffic, 2)
```

```python
office1_times = []
office2_times = []

for _, row in df.iterrows():
    home = row["Home Address"]
    o1 = row["Office 1 Address"]
    o2 = row["Office 2 Address"]

    office1_times.append(estimate_commute(home, o1))
    office2_times.append(estimate_commute(home, o2))

df["Office1_Time"] = office1_times
df["Office2_Time"] = office2_times

df
```

```python
df["Delta"] = df["Office2_Time"] - df["Office1_Time"]

df
```

```python
print("Average Office 1 commute:", df["Office1_Time"].mean())
print("Average Office 2 commute:", df["Office2_Time"].mean())

best_office = "Office 1" if df["Office1_Time"].mean() < df["Office2_Time"].mean() else "Office 2"
print("Better office (lower commute):", best_office)
```

```python
df.to_excel("commute_final_output.xlsx", index=False)
```

```python
from openpyxl import load_workbook
from openpyxl.styles import PatternFill
```

```python
wb = load_workbook("commute_final_output.xlsx")
ws = wb.active

green = PatternFill(start_color="C6EFCE", fill_type="solid")
red = PatternFill(start_color="FFC7CE", fill_type="solid")

# find Delta column index
headers = [cell.value for cell in ws[1]]
delta_col = headers.index("Delta") + 1

for row in ws.iter_rows(min_row=2):
    cell = row[delta_col - 1]

    if cell.value is not None:
        if cell.value >= 0:
            cell.fill = green
        else:
            cell.fill = red

wb.save("commute_final_output.xlsx")
```
